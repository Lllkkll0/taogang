```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l

def synthetic_data(w,b,num_exaples):
    """ y = Xw + b + """
    X = torch.normal(0,1,(num_exaples,len(w)))
    y = torch.matmul(X,w) + b
    y += torch.normal(0,0.01,y.shape)
    return X, y.reshape((-1,1))

true_w = torch.tensor([2,-3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
print('features:',features[0],'\nlabel:',labels[0])

d2l.set_figsize()
d2l.plt.scatter(features[:,(1)].detach().numpy(),labels.detach().numpy(),1)   

def data_iter(batch_size,features,labels):
    num_examples = len(features)  # 
    indices = list(range(num_examples)) # 
    # 
    random.shuffle(indices) # 
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(indices[i:min(i+batch_size,num_examples)]) # i+batch_sizenum_examples         
        yield features[batch_indices], labels[batch_indices] # 
        
batch_size = 10
for X,y in data_iter(batch_size, features, labels):
    print(X, '\n', y) # break
    break

# 
w = torch.normal(0,0.01,size=(2,1),requires_grad=True)
b = torch.zeros(1,requires_grad=True)

# 
def linreg(X,w,b):
    """"""
    return torch.matmul(X,w)+b

# 
def squared_loss(y_hat,y): #y_haty
    """"""
    return (y_hat - y.reshape(y_hat.shape))**2/2 # yy_hat  # 222 

# 
def sgd(params,lr,batch_size):#paramslrbatch_size
    """"""
    with torch.no_grad(): # 
        for param in params: # 
            param -= lr * param.grad / batch_size #  batch_size loss                          
            param.grad.zero_() # 

# 
lr = 0.03
num_epochs = 3#
net = linreg # net
loss = squared_loss  # 

# 
for epoch in range(num_epochs):
    for X,y in data_iter(batch_size,features,labels):
        l = loss(net(X,w,b),y) # xy
        # l(batch_size,1)l
        # [w,b]
        l.sum().backward()
        sgd([w,b],lr,batch_size) #
    with torch.no_grad():
        train_l = loss(net(features,w,b),labels)
        print(f'epoch{epoch+1},loss{float(train_l.mean()):f}')   

# 
print(f'w{true_w-w.reshape(true_w.shape)}')
print(f'b{true_b-b}')
```

```python
import numpy as np
from sklearn.linear_model import SGDRegressor
import time

# 1. 
np.random.seed(42)
n = 1000  # 
d = 5    # 

#  (1)
X = np.random.randn(n, d)
# 
true_weights = np.random.randn(d)
#  ()
y = X @ true_weights + np.random.randn(n) * 0.1

# 2. 
def linear_regression_analytic(X, y):
    """
    
    :
        X:  (n  d)
        y:  (n  1)
    
    :
        w:  (d  1)
    """
    #  X^T X
    XTX = X.T @ X
    
    # 
    if np.linalg.cond(XTX) > 1e12:
        print(": ")
    
    try:
        #  w = (X^T X)^{-1} X^T y
        w = np.linalg.inv(XTX) @ (X.T @ y)
    except np.linalg.LinAlgError:
        # 
        print("")
        w = np.linalg.pinv(XTX) @ (X.T @ y)
    
    return w

# 3. 
start_time = time.time()
w_analytic = linear_regression_analytic(X, y)
analytic_time = time.time() - start_time

# 4. SGD
start_time = time.time()
sgd = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.01)
sgd.fit(X, y)
sgd_time = time.time() - start_time

# 5. 
print("\n===  ===")
print(f": {true_weights}")
print(f": {w_analytic}")
print(f"SGD: {sgd.coef_}")
print(f"\n: {analytic_time:.6f}")
print(f"SGD: {sgd_time:.6f}")

# 6. 
analytic_error = np.mean((y - X @ w_analytic) ** 2)
sgd_error = np.mean((y - sgd.predict(X)) ** 2)
print(f"\n: {analytic_error:.6f}")
print(f"SGD: {sgd_error:.6f}")

# 7. 
def test_performance(n_samples, n_features):
    """"""
    X_test = np.random.randn(n_samples, n_features)
    true_weights = np.random.randn(n_features)
    y_test = X_test @ true_weights + np.random.randn(n_samples) * 0.1
    
    # 
    start = time.time()
    try:
        w_analytic = linear_regression_analytic(X_test, y_test)
    except:
        w_analytic = None
    analytic_time = time.time() - start
    
    # SGD
    start = time.time()
    sgd = SGDRegressor(max_iter=1000 if n_samples > 10000 else 100)
    sgd.fit(X_test, y_test)
    sgd_time = time.time() - start
    
    return analytic_time, sgd_time

# 
print("\n===  ===")
print(" (n=100, d=5):")
a_time, s_time = test_performance(100, 5)
print(f"  : {a_time:.6f}, SGD: {s_time:.6f}")

print("\n (n=10,000, d=100):")
a_time, s_time = test_performance(10000, 100)
print(f"  : {a_time:.6f}, SGD: {s_time:.6f}")

print("\n (n=1,000, d=2,000):")
a_time, s_time = test_performance(1000, 2000)
print(f"  : {a_time:.6f}, SGD: {s_time:.6f}")

print("\n (n=100,000, d=50):")
a_time, s_time = test_performance(100000, 50)
print(f"  : {a_time:.6f}, SGD: {s_time:.6f}")
```

```python
import hashlib
import os
import tarfile
import zipfile
import requests
import numpy as np
import pandas as pd
import torch
from torch import nn
from d2l import torch as d2l
%matplotlib inline

DATA_HUB = dict()
DATA_URL = 'http://d2l-data.s3-accelerate.amazonaws.com/'

def download(name, cache_dir=os.path.join('.', '01_data/02_DataSet_Kaggle_House')):
    """DATA_HUB"""
    assert name in DATA_HUB, f"{name}  {DATA_HUB}."
    url , sha1_hash = DATA_HUB[name]
    os.makedirs(cache_dir, exist_ok=True)
    fname = os.path.join(cache_dir, url.split('/')[-1])
    if os.path.exists(fname):
        sha1 = hashlib.sha1()
        with open(fname,'rb') as f:
            while True:
                data = f.read(1048576)
                if not data:
                    break
                sha1.update(data)
        if sha1.hexdigest() == sha1_hash:
            return fname
    print(f'{url}{fname}...')
    r = requests.get(url,stream=True,verify=True)
    with open(fname,'wb') as f:
        f.write(r.content)
    return fname
    
def download_extract(name, folder=None):
    """zip/tar"""
    fname = download(name)
    base_dir = os.path.dirname(fname)
    data_dir, ext = os.path.splitext(fname)
    if ext == '.zip':
        fp = zipfile.ZipFile(fname, 'r')
    elif ext in ('.tar', '.gz'):
        fp = tarfile.open(fname, 'r')
    else:
        assert False, 'zip/tar'
    fp,extractall(base_dir)
    return os.path.join(base_dir, folder) if folder else data_dir

def download_all():
    """DATA_UHB"""
    for name in DATA_HUB:
        download(name)

DATA_HUB['kaggle_house_train'] = (DATA_URL + 'kaggle_house_pred_train.csv','585e9cc9370b9160e7921475fbcd7d31219ce')         
DATA_HUB['kaggle_house_test'] = (DATA_URL + 'kaggle_house_pred_test.csv', 'fal9780a7b011d9b009e8bff8e99922a8ee2eb90')     
train_data = pd.read_csv(download('kaggle_house_train'))
test_data = pd.read_csv(download('kaggle_house_test'))
print(train_data.shape) # 146080te1label
print(test_data.shape) # label

print(train_data.iloc[0:4,[0,1,2,3,-3,-2,-1]]) # 
```

```python
import torch
from torch import nn

class ParallelBlock(nn.Module):
    def __init__(self, net1, net2):
        super().__init__()
        self.net1 = net1
        self.net2 = net2
    
    def forward(self, x):
        out1 = self.net1(x)
        out2 = self.net2(x)
        # 
        return torch.cat((out1, out2), dim=1)
    
net1 = nn.Linear(10, 5)
net2 = nn.Linear(10, 3)
parallel_net = ParallelBlock(net1, net2)

x = torch.randn(2, 10)
output = parallel_net(x)  # : (2, 8)
```

```python
def try_gpu(i=0):  #@save
    """gpu(i)cpu()"""
    if torch.cuda.device_count() >= i + 1:
        return torch.device(f'cuda:{i}')
    return torch.device('cpu')

def try_all_gpus():  #@save
    """GPUGPU[cpu(),]"""
    devices = [torch.device(f'cuda:{i}')
             for i in range(torch.cuda.device_count())]
    return devices if devices else [torch.device('cpu')]

try_gpu(), try_gpu(10), try_all_gpus()
```

```python
import torch
from torch import nn
from d2l import torch as d2l

def nin_block(in_channels, out_channels, kernel_size, strides, padding):
    return nn.Sequential(
        nn.Conv2d(in_channels, out_channels, kernel_size, strides, padding),
        nn.ReLU(),
        nn.Conv2d(out_channels, out_channels, kernel_size=1),
        nn.ReLU(),
        nn.Conv2d(out_channels, out_channels, kernel_size=1),
        nn.ReLU()
    )

def train_with_hyperparameters(lr=0.1, batch_size=128, dropout=0.5, epochs=10, 
                              optimizer_type='sgd', augmentation=False):
    # 
    net = nn.Sequential(
        nin_block(1, 96, kernel_size=11, strides=4, padding=0),
        nn.MaxPool2d(3, stride=2),
        nin_block(96, 256, kernel_size=5, strides=1, padding=2),
        nn.MaxPool2d(3, stride=2),
        nin_block(256, 384, kernel_size=3, strides=1, padding=1),
        nn.MaxPool2d(3, stride=2),
        nn.Dropout(dropout),
        nin_block(384, 10, kernel_size=3, strides=1, padding=1),
        nn.AdaptiveAvgPool2d((1, 1)),
        nn.Flatten()
    )
    
    # 
    from torchvision import transforms

    transform = [transforms.Resize(224)]
    if augmentation:
        transform.extend([
            transforms.RandomCrop(224, padding=10),
            transforms.RandomHorizontalFlip(),
        ])
    transform.append(transforms.ToTensor())
    transform = transforms.Compose(transform)
    
    train_iter, test_iter = d2l.load_data_fashion_mnist(
        batch_size, resize=224)
    
    # 
    device = d2l.try_gpu()
    net.to(device)
    if optimizer_type == 'adam':
        optimizer = torch.optim.Adam(net.parameters(), lr=lr)
    else:
        optimizer = torch.optim.SGD(net.parameters(), lr=lr, momentum=0.9)
    
    # 
    print(f": lr={lr}, batch_size={batch_size}, dropout={dropout}, "
          f"epochs={epochs}, optimizer={optimizer_type}, "
          f"augmentation={augmentation}")
    
    results = d2l.train_ch6(net, train_iter, test_iter, epochs, lr, device, 
                           optimizer=optimizer, verbose=True)
    
    print(f": {results[0]:.2f}%")
    return results[0]

# 1: 
print("===== 1:  =====")
print("\n ():")
baseline_acc = train_with_hyperparameters()

print("\n ():")
optimized_acc = train_with_hyperparameters(
    lr=0.001, 
    batch_size=256, 
    dropout=0.3, 
    epochs=15, 
    optimizer_type='adam',
    augmentation=True
)

print(f"\n:")
print(f": {baseline_acc:.2f}%")
print(f": {optimized_acc:.2f}%")
print(f": {optimized_acc - baseline_acc:.2f}%")
```

```python
import torch
from torch import nn
from d2l import torch as d2l


def batch_norm(X, gamma, beta, moving_mean, moving_var, eps, momentum):
    # is_grad_enabled
    if not torch.is_grad_enabled():
        # 
        X_hat = (X - moving_mean) / torch.sqrt(moving_var + eps)
    else:
        assert len(X.shape) in (2, 4)
        if len(X.shape) == 2:
            # 
            mean = X.mean(dim=0)
            var = ((X - mean) ** 2).mean(dim=0)
        else:
            # axis=1
            # X
            mean = X.mean(dim=(0, 2, 3), keepdim=True)
            var = ((X - mean) ** 2).mean(dim=(0, 2, 3), keepdim=True)
        # 
        X_hat = (X - mean) / torch.sqrt(var + eps)
        # 
        moving_mean = momentum * moving_mean + (1.0 - momentum) * mean
        moving_var = momentum * moving_var + (1.0 - momentum) * var
    Y = gamma * X_hat + beta  # 
    return Y, moving_mean.data, moving_var.data
```

```python
class BatchNorm(nn.Module):
    # num_features
    # num_dims24
    def __init__(self, num_features, num_dims):
        super().__init__()
        if num_dims == 2:
            shape = (1, num_features)
        else:
            shape = (1, num_features, 1, 1)
        # 10
        self.gamma = nn.Parameter(torch.ones(shape))
        self.beta = nn.Parameter(torch.zeros(shape))
        # 01
        self.moving_mean = torch.zeros(shape)
        self.moving_var = torch.ones(shape)

    def forward(self, X):
        # Xmoving_meanmoving_var
        # X
        if self.moving_mean.device != X.device:
            self.moving_mean = self.moving_mean.to(X.device)
            self.moving_var = self.moving_var.to(X.device)
        # moving_meanmoving_var
        Y, self.moving_mean, self.moving_var = batch_norm(
            X, self.gamma, self.beta, self.moving_mean,
            self.moving_var, eps=1e-5, momentum=0.9)
        return Y
```

```python
net = nn.Sequential(
    nn.Conv2d(1, 6, kernel_size=5), BatchNorm(6, num_dims=4), nn.Sigmoid(),
    nn.AvgPool2d(kernel_size=2, stride=2),
    nn.Conv2d(6, 16, kernel_size=5), BatchNorm(16, num_dims=4), nn.Sigmoid(),
    nn.AvgPool2d(kernel_size=2, stride=2), nn.Flatten(),
    nn.Linear(16*4*4, 120), BatchNorm(120, num_dims=2), nn.Sigmoid(),
    nn.Linear(120, 84), BatchNorm(84, num_dims=2), nn.Sigmoid(),
    nn.Linear(84, 10))
```

```python
lr, num_epochs, batch_size = 1.0, 10, 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

width, height = 200, 200  # 
r = 20  # 
x = r + 20  # 
y = r + 100  # 
x_offer = y_offer = 4  # 

fig, ax = plt.subplots()
ax.set_xlim(0, width)
ax.set_ylim(0, height)
circle = plt.Circle((x, y), r, color='b')
ax.add_patch(circle)
ax.set_aspect('equal')
plt.gca().invert_yaxis()  # yOpenCV

def animate(frame):
    global x, y, x_offer, y_offer
    if x > width - r or x < r:
        x_offer *= -1
    if y > height - r or y < r:
        y_offer *= -1
    x += x_offer
    y += y_offer
    circle.center = (x, y)
    return (circle,)

ani = animation.FuncAnimation(fig, animate, frames=300, interval=16, blit=True)
plt.show()
```
